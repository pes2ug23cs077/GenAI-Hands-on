```python
import os
import getpass
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 2. Securely Set API Key
if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("Enter your Google API Key: ")

# 3. Initialize Gemini Model
# If this model gives 404, try "gemini-2.0-pro"
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)

# 4. One-Line LCEL Chain
chain = (
    ChatPromptTemplate.from_template(
        "What year was {movie} released? Answer with only the 4-digit year."
    )
    | llm
    | StrOutputParser()
    | (lambda year: 2026 - int(year.strip()))
)

# 5. Run Test
movie_name = "Inception"
years_ago = chain.invoke({"movie": movie_name})

print(f"'{movie_name}' was released {years_ago} years ago.")
```
